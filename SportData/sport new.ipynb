```python
import pandas as pd
urls = [f"https://www.skysports.com/premier-league-table/{year}" for year in range(2011,2021)]
all_datas = []
for url in urls:
    try:
        all_datas.append(pd.read_html(url)[0])
        
    except Exception as err:
        print(err)
```
